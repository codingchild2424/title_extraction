```python
import json
import os
import pandas as pd

data_path = "/workspace/home/uglee/Projects/title_extraction/datasets/raw_datasets/ai_hub_essay_json"

for idx, json_file_name in enumerate(os.listdir(data_path)):

    try:
        json_file_path = os.path.join(data_path, json_file_name)

        json_string = open(json_file_path, encoding='utf-8')

        #print("json_string", json_string)

        json_data = json.load(json_string)

        title = json_data['rubric']['essay_main_subject']
        content = json_data['paragraph'][0]['paragraph_txt']

        json_string.close()

        # 전처리
        title = title.replace("\t", "")
        title = title.replace("\n", " ")
        title = title.replace("|", "")

        content = content.replace("\t", "")
        content = content.replace("\n", " ")
        content = content.replace("|", "")

        title_content = title + "\t" + content

        tsv_file_name = json_file_name.split('.')[0] + ".tsv"

        tsv_file_path = os.path.join(data_path, tsv_file_name)

        f = open(tsv_file_path, 'w', encoding='utf-8')

        f.write(title_content)

        f.close()
    except:
        continue

```
