```python
import os

gangwon_path = "/workspace/home/uglee/Projects/title_extraction/datasets/raw_datasets/gangwon_folder"

gangwon_file_list = os.listdir(gangwon_path)
```

```python
# 파일 하나씩 읽기

pre_path = "/workspace/home/uglee/Projects/title_extraction/datasets/pre_datasets"

for file_name in gangwon_file_list:
    file_path = os.path.join(gangwon_path, file_name)

    print("file_path", file_path)

    f = open(file_path, 'r', encoding='utf-8')

    text = f.read()

    f.close()

    clean_text_list = text.split('\n')

    title = clean_text_list[0]
    content = '\n'.join(clean_text_list[1:])

    # 데이터 전처리
    title = title.replace("\t", "")
    title = title.replace("\n", " ")

    content = content.replace("\t", "")
    content = content.replace("\n", " ")

    title_content = title + "\t" + content

    tsv_path = os.path.join(pre_path, title + ".tsv")

    f2 = open(tsv_path, 'w', encoding='utf-8')

    f2.write(title_content)

    f2.close()
```
